The primary goal of this project is to analyze a dataset of restaurants to gain insights into their distribution across various price range categories. This involves performing data preprocessing, exploratory data analysis, and visual interpretation using Python (likely with libraries like pandas, matplotlib, or seaborn).

Objectives:

Import and clean the dataset to ensure accurate analysis.

Categorize restaurants based on price indicators (e.g., price range levels from 1 to 4).

Visualize the price range distribution using a bar chart or histogram.

Calculate the percentage of restaurants falling into each price category.

This analysis helps identify market trends, consumer affordability preferences, and potential opportunities for new business entries in different pricing segments.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv(r"C:\Users\admin\Downloads\Dataset  (1).csv")

# Step 1: Analyze rating distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['Aggregate rating'], bins=20, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Step 2: Most common rating
most_common_rating = df['Aggregate rating'].mode().iloc[0]
print(f"Most common rating: {most_common_rating}")

# Step 3: Average number of votes
average_votes = df['Votes'].mean()
print(f"Average number of votes: {round(average_votes, 2)}")
```
