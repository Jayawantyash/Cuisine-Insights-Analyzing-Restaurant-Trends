Project Overview
This project investigates the relationship between restaurant price range and the availability of two key services:

Online Delivery

Table Booking

Using a restaurant dataset, the analysis groups restaurants by their price range and calculates the percentage that offers each service within those groups. The goal is to understand whether service offerings differ across low-cost and high-cost restaurants, which can inform business strategies for customer service enhancement and digital transformation.

The analysis involves:

Cleaning and filtering data to ensure completeness.

Aggregating service availability data by price range.

Visualizing the results using line plots for comparative insights.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(r"C:\Users\admin\Downloads\Dataset  (1).csv")  # replace with your actual file path

# Replace with actual column names
price_col = 'Price range'
online_col = 'Has Online delivery'
booking_col = 'Has Table booking'

# Drop rows with missing values in relevant columns
df = df.dropna(subset=[price_col, online_col, booking_col])

# Group by price range and calculate % of online delivery and table booking
grouped = df.groupby(price_col).agg({
    online_col: lambda x: (x == 'Yes').mean() * 100,
    booking_col: lambda x: (x == 'Yes').mean() * 100
}).reset_index()

# Rename columns for clarity
grouped.columns = ['Price Range', 'Online Delivery %', 'Table Booking %']

# Plotting
plt.figure(figsize=(10, 6))
sns.lineplot(data=grouped, x='Price Range', y='Online Delivery %', marker='o', label='Online Delivery')
sns.lineplot(data=grouped, x='Price Range', y='Table Booking %', marker='o', label='Table Booking')
plt.title('Service Availability by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Percentage of Restaurants Offering Service')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Print the data
print("\nPercentage of Online Delivery and Table Booking by Price Range:")
print(grouped)
```

Conclusion
The results of the analysis highlight the following insights:

Online Delivery tends to be more common among lower to mid-range price levels, indicating a demand for convenience in budget-friendly dining options.

Table Booking becomes more prevalent in higher-end restaurants, suggesting a focus on planned dining experiences and better service expectations.

This contrasting trend suggests that service offerings are aligned with the expectations of different customer segments based on spending habits.

These findings can guide restaurant businesses in tailoring their service models according to price segment, potentially boosting customer satisfaction and operational efficiency.
