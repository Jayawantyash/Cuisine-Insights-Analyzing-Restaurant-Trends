```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv(r"C:\Users\admin\Downloads\Dataset  (1).csv")

# Basic cleanup
df = df.dropna(subset=['Aggregate rating'])

# --- 1. Rating Distribution ---
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Aggregate rating', palette='viridis')
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Count')
plt.show()

# --- 2. Top Cities by Number of Restaurants ---
top_cities = df['City'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_cities.values, y=top_cities.index, palette='magma')
plt.title('Top 10 Cities with Most Restaurants')
plt.xlabel('Number of Restaurants')
plt.ylabel('City')
plt.show()

# --- 3. Most Common Cuisines ---
top_cuisines = df['Cuisines'].dropna().str.split(', ').explode().value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_cuisines.values, y=top_cuisines.index, palette='coolwarm')
plt.title('Top 10 Cuisines')
plt.xlabel('Frequency')
plt.ylabel('Cuisine')
plt.show()

# --- 4. Average Cost vs Rating ---
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Average Cost for two', y='Aggregate rating', alpha=0.6)
plt.title('Cost for Two vs Aggregate Rating')
plt.xlabel('Average Cost for Two')
plt.ylabel('Aggregate Rating')
plt.show()
```

Conclusion
The analysis provided meaningful insights into consumer preferences and market trends in the restaurant industry:

Ratings are not evenly distributed, suggesting customer experience varies significantly across establishments.

A few major cities dominate the restaurant scene, which could indicate higher demand or population concentration.

Popular cuisines tend to be globally recognizable, such as North Indian and Chinese, suggesting a trend toward widely accepted tastes.

There's no strong linear correlation between price and rating, indicating that higher prices do not necessarily translate to higher customer satisfaction.

These findings can support strategic decisions in areas like restaurant marketing, expansion planning, and menu design.
