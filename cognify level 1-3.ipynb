This project analyzes restaurant data with a focus on understanding how restaurants are distributed across different price range categories. The dataset includes various attributes of restaurants such as name, location, ratings, and price level.

The primary objectives of this analysis were:

Categorize restaurants by price range (e.g., low, medium, high, luxury).

Visualize the distribution of restaurants in each price category using a bar chart or histogram.

Calculate the percentage of restaurants in each category to better understand pricing trends in the dataset.

By doing so, the project aimed to uncover pricing patterns that could be valuable for consumers, restaurant owners, and market analysts.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(r"C:\Users\admin\Downloads\Dataset  (1).csv")

# Count and calculate percentage
price_counts = df['Price range'].value_counts().sort_index()
price_percentages = (price_counts / len(df)) * 100

# Plot
plt.figure(figsize=(8, 5))
price_percentages.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribution of Price Ranges Among Restaurants')
plt.xlabel('Price Range')
plt.ylabel('Percentage of Restaurants')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Print percentages
print(price_percentages.round(2))

```

From the analysis, we found that:

A significant portion of the restaurants fall within the mid-range price category.

Low-cost and high-end restaurants are less common, with luxury restaurants being the least represented.

The bar chart clearly illustrated the skewness of price distribution toward mid-range pricing.

Key Takeaways:

Mid-range pricing dominates the restaurant market, suggesting a strong demand for affordability without compromising on quality.
