This project focuses on geospatial analysis of restaurant data by mapping their locations using latitude and longitude. The dataset includes detailed information about restaurants such as name, city, address, cuisines, ratings, and coordinates.

Objectives:
Visualize restaurant locations on a scatter plot using their geographical coordinates.

Identify geographic clusters or concentrations of restaurants.

Compare distribution across cities to highlight areas with high or low restaurant density.

Tools Used:
Python for data analysis and visualization.

Pandas for data manipulation.

Seaborn & Matplotlib for plotting.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path =  (r"C:\Users\admin\Downloads\Dataset  (1).csv") # Update this path if necessary
df = pd.read_csv(file_path)

# Set up the plot style
sns.set(style="whitegrid")

# Plotting the locations of restaurants
plt.figure(figsize=(10, 8))
sns.scatterplot(
    data=df,
    x="Longitude",
    y="Latitude",
    hue="City",
    palette="tab10",
    legend='full',
    s=100
)

plt.title("Restaurant Locations by City")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend(title="City", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
